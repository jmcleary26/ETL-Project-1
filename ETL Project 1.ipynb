```python
##Starter Code for the cleaning of the Data
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Cost of Living Excel import
costofliving = "Resources/cost_of_living_index.csv"
col_data_df = pd.read_csv(costofliving)
col_data_df.head()
```

```python
#GlassDoor Data Import
glassdoor = "Resources/glassdoor.csv"
glassdoor_data_df = pd.read_csv(glassdoor)
glassdoor_data_df.head()
```

```python
costofliving_df = pd.DataFrame(col_data_df, columns = ["City", "Cost of Living Index"])
costofliving_df
```

```python
costofliving_df['City']=costofliving_df.City.str.split(',',expand=True)
costofliving_df
```

```python
glassdoor_df= pd.DataFrame(glassdoor_data_df, columns = ["Job Title", "Location", "Size", "Type of Ownership", "Industry", "avg_salary"])
glassdoor_df
```

```python
glassdoor_df[['Location']]=glassdoor_df['Location'].str.split(",",expand=True)[0]
glassdoor_df
```

```python
glassdoor_df["Job Title"].unique()
```

```python
glassdoor_df["avg_salary"].describe()
```

```python
min_df = glassdoor_df.loc[glassdoor_df['avg_salary'] == 13.5]
min_df
```

```python
bins = [0, 15, 40, 60, 80, 100, 200, 300]
labels = ["<15", "16-40", "41-60", "61-80", "81-100", "101-200", "201-300"]

# Slice the data and place it into bins
glassdoor_df["Salary Range"] = pd.cut(glassdoor_df["avg_salary"], bins, labels=labels, include_lowest=True)
glassdoor_df.head()
```

```python
grouped_by_salary = pd.DataFrame(glassdoor_df.groupby(["Salary Range"])["Job Title"].count()).rename(columns={"Job Title": "Job Count"})
grouped_by_salary.plot(kind="bar")
```
